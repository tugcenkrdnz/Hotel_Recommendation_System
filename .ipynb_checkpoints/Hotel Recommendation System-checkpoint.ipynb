**Projenin Amacı**

Bu projenin temel amacı, devasa boyuttaki otel veri setini (500.000+ yorum) kullanarak, kullanıcının hem konum hem de ihtiyaçlarına (niyetine) göre kişiselleştirilmiş otel önerileri sunan bir sistem geliştirmektir.

```python
# Verileri tablolar halinde düzenlemek, analiz etmek ve filtrelemek için kullanılır.
import pandas as pd

# Metinlerdeki kelimeleri, önem derecelerine göre matematiksel sayılara (vektörlere) dönüştürür.
from sklearn.feature_extraction.text import TfidfVectorizer

# Sayısal hale gelmiş metinler arasındaki benzerliği, aralarındaki açıyı ölçerek hesaplar.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#veri setini yüklüyoruz
df=pd.read_csv('Hotel_Reviews.csv')
```

# EDA

```python
#ilk 5 satırı göster
df.head()
```

```python
#veriler hakkında bilgi
df.info()
```

```python
#kaç satır kaç sütun
df.shape
```

```python
#Hızlı olması için karışık 100000 satır ile işlem yapacağız
# n=10000: Kaç satır alınacağı
# random_state=42: Her çalıştırdığında aynı "rastgele" listeyi almamızı sağlar
df = df.sample(n=100000, random_state=42).reset_index(drop=True)
```

```python
## Veri setinde bulunan 10 otelin adresini listeler.
df['Hotel_Address'].value_counts(10)
```

```python
# Adresin en sonundaki kelimeyi alıp 'Countries' adlı yeni sütuna yazıyoruz
df['Countries'] = df['Hotel_Address'].apply(lambda x: x.split()[-1])
```

```python
# Otellerin ülkelere göre dağılımını
df['Countries'].value_counts()
```

```python
# 'Kingdom' olanları 'UK' olarak düzelterek temizleyelim
df['Countries'] = df['Countries'].replace('Kingdom', 'UK')
```

```python
# Ortalama puanı (Average_Score) en yüksek olan benzersiz otelleri listeler
# drop_duplicates kullanarak aynı otelin tekrar listelenmesini engelliyoruz
df[['Hotel_Name', 'Average_Score']].drop_duplicates().sort_values(by='Average_Score', ascending=False).head(10)
```

```python
# Ülke dağılımını görsel bir grafik olarak bastırır
df['Countries'].value_counts().plot(kind='bar', title='Ülkelere Göre Otel Sayısı')
```

```python
# Gerekli sütunları seçiyoruz
df = df[['Hotel_Name', 'Tags', 'Positive_Review','Countries']].copy()
```

```python
# Öneri için kullanılacak metni oluştur: Etiketler ve Olumlu yorumları birleştiriyoruz
df['Features'] = df['Tags'] + " " + df['Positive_Review']
```

```python
def recommend_hotel(country, description):
    # 1. Tüm veri setini değil, sadece senin belirttiğin ülkedeki otelleri ayırır.
    sub_df = df[df['Countries'].str.contains(country, case=False)].copy()
    
    # 2. Metin analiz aracını (TF-IDF) hazırlar.
    tf = TfidfVectorizer(stop_words='english')
    
    # 3. O ülkedeki otellerin özelliklerini (Features) sayılara döker ve senin 
    # yazdığın cümleyle (description) ne kadar benzediğini (cosine_similarity) ölçer.
    skorlar = cosine_similarity(tf.fit_transform(sub_df['Features']), tf.transform([description])).flatten()
    
    # 4. Benzerlik puanlarını büyükten küçüğe sıralar ve en yüksek puanlı 3 otel ismini 
    # sade bir liste (.values) olarak döndürür.
    return sub_df['Hotel_Name'].iloc[skorlar.argsort()[-3:][::-1]].values
```

```python
#öneri: isteğine en uygun, en benzer 3 oteli ekrana getirir.
recommend_hotel('Austria', 'I want a hotel with breakfast')
```

```python
#öneri: isteğine en uygun, en benzer 3 oteli ekrana getirir.
recommend_hotel('UK', 'I am going for business trip')
```

**Projenin Sonucu**

Proje başarıyla tamamlanmış ve aşağıdaki çıktılar elde edilmiştir:

Akıllı Arama Motoru: Klasik anahtar kelime aramasının ötesinde, metin benzerliği (Cosine Similarity) kullanarak kullanıcının niyetine en yakın otelleri bulan bir algoritma oluşturulmuştur.

Verimli Veri Yönetimi: 100.000 satırlık karma (sample) bir veri seti üzerinde, "United Kingdom" gibi karmaşık adres yapılarından ülke bilgisini ayrıştıran temiz bir veri ön işleme (preprocessing) mimarisi kurulmuştur.

Fonksiyonel Çözüm: recommend_hotel fonksiyonu ile tek bir satır kod yazarak, herhangi bir ülkeden en alakalı 3 otel tavsiyesini saniyeler içinde getiren kullanıcı dostu bir araç geliştirilmiştir.
